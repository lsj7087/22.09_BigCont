**loan_result**의 "application_id"와 **user_spec**의 "application_id"를 기준으로 merge

```python
tmp = pd.merge(loan_result[-loan_result.duplicated()],user_spec, how='left',on='application_id')
```

```python
tmp
```

```python
# 이 중에서 user_id가 결측인 값 => loan_result에 있는 application_id가 user_spec에는 없었다. 113개
# user에 대한 정보를 얻을 수 없다면 제거하는게 적절하다고 봄.
tmp[tmp["user_id"].isna()]
```

```python
# 113개 전부 대출 신청한 상품임
tmp[tmp["user_id"].isna()]["is_applied"].sum()
```

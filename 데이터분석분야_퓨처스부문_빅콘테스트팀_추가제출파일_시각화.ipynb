## income_type

```python
sum_income_type_day = user_spec.groupby('income_type')["application_id"].count()
# sum_income_type_day = sum_income_type_day.sort_values(ascending=True)
ratio_sum_income_type_day = sum_income_type_day/sum_income_type_day.sum()
c = sns.color_palette("Blues")

x_label = ratio_sum_income_type_day.index.tolist()
f = plt.figure(figsize=[10,6])

plt.pie(ratio_sum_income_type_day, # 비율 값
        #labels=x_label, # 라벨 값
        #autopct='%.1f%%', # 부채꼴 안에 표시될 숫자 형식(소수점 1자리까지 표시)
        startangle=90, # 축이 시작되는 각도 설정
        counterclock=True, # True: 시계방향순 , False:반시계방향순
        explode=[0.05,0.05,0.05,0.05,0.05,0.05], # 중심에서 벗어나는 정도 표시
        shadow=True, # 그림자 표시 여부
        colors = [c[5], c[4], c[3], c[2], c[1], c[0]],
        wedgeprops = {'width':0.7,'edgecolor':'w','linewidth':1}
        ) #width: 부채꼴 영역 너비,edgecolor: 테두리 색 , linewidth : 라인 두께
plt.title('Ratio of income_type', fontsize=16)
plt.show()
```
